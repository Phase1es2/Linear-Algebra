**Exercise 3**. We want to predict ```Education``` from the other variables (not including ```Providence```). Fit the following regression model to the data:

\begin{align} \hat{Education} = \hat{\beta_0} + \hat{\beta_1}Fertility + \hat{\beta_2}Agriculture + \hat{\beta_3}Examination + \hat{\beta_4}Catholic + \hat{\beta_5}Infant\_Mortality.\end{align}

$\hat{\beta_4}$ = ____. 



````python
import numpy as np

X = swisses.to_numpy()[:,[1,2,3,5,6]]# Note that we don't need to do a column of ones since sklearn will do it for us.
y = swisses.to_numpy()[:,4]
print(X)
print()
print(y)
````

````python
from sklearn.linear_model import LinearRegression

swiss_model = LinearRegression().fit(X, y)

print(swiss_model.coef_)
print()
print(swiss_model.intercept_)

#Q3 0.1002
````

**Exercise 4**. The model from the previous exercise predicts that a providence with ```Fertility``` = 80.2, ```Agriculture``` = 17, ```Examination``` = 15, ```Catholic``` = 9.96, and ```Infant_Mortality``` = 22.2 will have a post-primary military draftee education level of ___%.

````python
swiss_in = np.array([[80.2,17,15,9.96,22.2]])

print(swiss_model.predict(swiss_in))
#9.0469
````

**Exercise 5**. This model explains ____% of the variation in the ```Education``` variable.

````python
#R2
r2 = swiss_model.score(X, y)
print("R^2:", r2)
#0.7678
````

**Exercise 6**. Find the Residual Squared Error (RSE) for our model. (Hint: Find SSE by first finding MSE.) 

````python
import math
y_hat = swiss_model.predict(X)
print("y_hat", y_hat)
sse = np.sum((y-y_hat)**2)
print("sse", sse)
n = X.shape[0]
mse = sse / n
print("MSE", mse)
p = X.shape[1]
rse = math.sqrt(sse / (n - p - 1))
print("RSE", rse)
#4.9073
````

**Exercise 7**. Consider every possible least squares regression model to predict ```Education``` that uses only a single variable as its input. The model that performs the best, as measured by MSE, is \begin{align}\hat{Education} = \hat{\beta_0} + \hat{\beta_1}x, \end{align} where x = "_______".

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = swisses.to_numpy()[:,[1]]
X1 = swisses.to_numpy()[:,[2]]
#X = fertility.reshape(-1, 1)
#X1 = agriculture.reshape(-1,1)
y = swisses['Education'].to_numpy()

fertility_model = LinearRegression().fit(X, y)
print(fertility_model.coef_)

print(fertility_model.intercept_)
print()
agriculture_model = LinearRegression().fit(X1, y)
print(agriculture_model.coef_)
print(agriculture_model.intercept_)


y_pred = fertility_model.predict(X)
mse = mean_squared_error(y, y_pred)


y1_pred = agriculture_model.predict(X1)
mse1 = mean_squared_error(y, y1_pred)

print("mse", mse)
print("mse1", mse1)
````
