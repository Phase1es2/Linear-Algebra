````python
# Step 4: Preprocess the dataset by standardizing each variable
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
x = scaler.fit_transform(x)
print(x)
````

````python
from sklearn import cluster
import numpy as np
import matplotlib.pyplot as plt
# Step 5: Initialize K-Means cluster object
kmeans_iris = cluster.KMeans(n_clusters = 3, n_init = 10, random_state = 0)

# Step 6: Use kmeans_iris's "fit_predict()" method to predict the groups for each value in the dataset
label = kmeans_iris.fit_predict(x)

# Step 7: Get unique labels (group labels) using numpy's "unique" method
u_labels = np.unique(label)

# Step 8: Plot the results:
centers = np.array(kmeans_iris.cluster_centers_) 
for i in u_labels:
    plt.scatter(x[label == i , 0] , x[label == i , 1], label = i)
plt.scatter(centers[:,0], centers[:,1], marker="x", color='k')
plt.legend()
plt.show()
````

Something you probably noticed after doing this example is that the graph showing the K-Means' groupings do not appear very "clustered." There are two possible reasons for this.

### III. Principle Component Analysis

Principle Component Analysis (PCA) is a *dimension reduction* techniqe. The mathematics (described in Chapter 15 of our textbook) are fairly complex, but the idea is straightforward: project a vector (or data point) from a higher dimensional space to a lower dimensional space. More specifically, the direction of largest variation (equivalently, largest standard deviation) becomes the "first principal componenet" of the dataset. The second principal component is found by considering all possible directions that are orthogonal to the first principal component and then choosing the direction that has the highest variance. And so on. For visualization purposes it is common to just use the first two principle components (can you think of why this would be?).

**Example 1**. Let's use the ```PCA``` method from the ```decomposition``` module (which is in the ```sklearn``` libary) to reduce the ```wine``` dataset to 2 dimensions. Then we'll run the K-Means algorithm on these two dimensions. 

````python
from sklearn.decomposition import PCA

pca = PCA(2) # initialize the PCA for reducing to 2 dimensions
x = pca.fit_transform(x) # run the PCA dimension reduction on the wines dataset (assumes that X was the dataset you produced in step 4)
````

````python
from sklearn import cluster
import matplotlib.pyplot as plt

# Now that you've reduced the data's dimensions using the PCA, run the K-means as usual

kmeans_wine_pca = cluster.KMeans(n_clusters=3, n_init = 10, random_state=0)
label = kmeans_wine_pca.fit_predict(x)
u_labels = np.unique(label)
  
for i in u_labels:
    plt.scatter(x[label == i , 0] , x[label == i , 1], label = i)
plt.legend()
plt.show()
````

Notice that running K-Means after using the PCA to reduce to 2 dimensions gives us much more distinct groupings (with no apparent group overlaps) than before.
