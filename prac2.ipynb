```python
from sklearn import preprocessing
X = t.to_numpy()[:,[1,2,3]]

y = t.to_numpy()[:,0]
print(y)
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler() #construct the scalar object
X_std = scaler.fit_transform(X) #deploy and calculate 
rows, columns = X_std.shape
print(rows)
print(columns)

print(X_std)
```

```python
from sklearn.linear_model import LogisticRegression

t_model = LogisticRegression().fit(X_std, y)

print(t_model.coef_)
print()
print(t_model.intercept_)
```

```python
from sklearn import metrics

y_hat = t_model.predict(X_std)
confusion_matrix = metrics.confusion_matrix(y,y_hat, labels = t_model.classes_)

print(confusion_matrix)
```

```python
tn, fp, fn, tp = confusion_matrix.ravel()
accuracy = (tp + tn)/X.shape[0]
print(accuracy)
#0.8114
```

```python
precision = tp/(tp+fp)
recall = tp/(tp + fn)

f_1 = 2*precision*recall/(precision + recall)
print(f_1)

#0.7423
```
