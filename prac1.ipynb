```python
import numpy as np

#A = np.array([[1,1,1],[2,1,0],[-2,-1,0],[1,0,1]])
#B = np.array([[2],[5],[4],[1]])
A = np.array([[1,1,1],[2,1,0],[-2,-1,0],[1,0,1]])
B = np.array([[2],[-2],[4],[-4]])
A_T_A = A.T@A
print(A_T_A)
print()
A_in = np.linalg.inv(A_T_A)
print(A_in)
print()
x = A_in@A.T@B

print(x)

print()
axb = A@x-B
print(axb)

print(np.linalg.norm(axb))
```
