### I. Matrix Inverse

A good starting place is to think about solution(s) to $Ax = b$.

To solve this equation you have 3 options:

1. Solve an equation for a variable, then backsolve (high school algebra).  
2. Set up an augmented matrix $[A | b]$ and perform elementary row operations.
3. Find the matrix inverse $A^{-1}$ and express $x$ as the matrix-vector product $A^{-1}b$.

#### Conditions for a matrix $A$ to have a full-inverse $A^{-1}$:

A <u>full-inverse</u> of $A$ is a matrix $A^{-1}$ such that $AA^{-1} = I$ and $A^{-1}A = I$.
//only square matrix
1. $A$ must have $M = N$.
2. $A$ must be "full-rank", ie, $r = M = N$.

#### II. One-Sided Inverses

A <u>left-inverse</u> is a matrix $L$ such that $LA = I$.

A <u>right-inverse</u> is a matrix $R$ such that $AR = I$.
