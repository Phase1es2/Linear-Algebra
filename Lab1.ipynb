Read pages 297-300 in *Practical Linear Algebra*. If you have seen "if" statements and "for loops" before this material will be very easy for you; if not be sure to carefully review the assigned reading and spend some time playing with these examples. We will see Control Flow in class next week. 

**Example 1**. Consider the following for-loop:

````python
n = 10
for i in range(n):
  print(i)

print("Done.")
````

```range(n)``` is an iterable generator function, it produces the "next" number in a sequence on demand. The first thing that happens in the above code block is that we assign the integer 10 to the variable name "n". Then the for-loop runs: *i* is a placeholder variable that changes (increments by +1) every time we go through the loop. The first value that range(10) produces is 0 (by default). Then print(0) happens, then control is returned to the top of the loop. range(10) produces 1, which is then passed to the body of the loop and print(1) happens, and so on. The 10th value produced by range(10) is the last value that is run, which is 9; print(9) happens, then control is passed to the next line after the for-loop, which is print("Done.")
