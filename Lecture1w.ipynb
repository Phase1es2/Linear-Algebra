### III. Calculus and Linear Algebra

**Example**: $f(x) = x^2 -2$. Describe this function using a graph, sets, and its tangent line function. 

Calculus uses tools from algebra and geometry to study the continuous way (rate) the outputs of a function change as the inputs change. Linear Algebra (on the other hand) is the study of vectors and systems of linear equations.

### IV. Data Science

A *model* is a mathematical description of the relationship between two or more different variables. This kind of modeling is common in statistics, machine learning, dynamical systems, regression, and classification. 

*Data science* is the process of using data to formulate a model that gives us insight about an underlying phenomena that we wish to study. Modern data science involves tools from statistics, computer science, machine learning, and other areas of mathematics (depnding on the specific application). Linear Algebra lies at the heart of data science and modeling applications, which is why we study it specifically in this course.
